```python
# Let's split the data to train and test
from sklearn.model_selection import StratifiedShuffleSplit

#first we create a category for the median income, so that when we split, we're garanted to obtain a representative test set

housing_data["income_cut"] = pd.cut(housing_data["median_income"],
                                    bins=[0., 1.5, 3.0, 4.5, 6., np.inf],
                                    labels=[1, 2, 3, 4, 5])

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)

for train_index, test_index in split.split(housing_data, housing_data["income_cut"]):
    strat_train_set = housing_data.loc[train_index]
    strat_test_set  = housing_data.loc[test_index]  
```

```python
# we check if the test contains what we expect: 
print("test_set income categories \n", strat_test_set["income_cut"].value_counts() / len(strat_test_set))
# we check if the full data contains what we expect: 
print("test_set income categories \n", housing_data["income_cut"].value_counts() / len(housing_data))
```

```python
# we remove our income_cut , now that we don't need it 

for set_ in (strat_train_set, strat_test_set):
    set_.drop("income_cut", axis=1, inplace=True)
```

```python
# we have a very nicely balanced trainning and test data
# the next step is to start our analysis
housing_data.plot(kind="scatter", x="longitude", y="latitude", alpha=0.1)
```

```python
# very nice, we can clearly see the locations with the biggest densities
# but what about the prices ???
# we could give it some colors !!! 

housing_data.plot(kind="scatter", x="longitude", y="latitude", alpha=0.4,
            s=housing_data["population"]/100, label="population", figsize=(10,7),
            c="median_house_value", cmap=plt.cm.get_cmap("jet"), colorbar=True)
```
